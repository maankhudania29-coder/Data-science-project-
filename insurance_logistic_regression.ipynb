```python
import pandas as pd
data= pd.read_csv('New_insurance.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split#(for traing and testing)
from sklearn.linear_model import LogisticRegression#(for prediction)
from sklearn.preprocessing import StandardScaler#(for filtering)
from imblearn.over_sampling import SMOTE#(over evaluation for minority data)
import numpy as np#(for numerical data handling)
```

```python
X = data.drop(['charges','insuranceclaim'], axis =1)
y = data['insuranceclaim']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size= 0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
print("Insurance claim Prediction")
age = float(input("Enter age in years: "))
sex = int(input("Enter gender, 0 = female, 1 = male: "))
bmi = float(input("Enter bmi: "))
children = int(input("Enter number of children: "))
smoker = int(input("Enter smoking status, 0 = No, 1 = Yes: "))
region = int(input("Enter region: 0 = Northeast, 1 = northwest, 2 = southeast, 4 = southwest: "))
user_input = pd.DataFrame([[age, sex, bmi, children, smoker, region]], columns=['age', 'sex', 'bmi', 'children', 'smoker', 'region'])  # Fixed: column= -> columns= and fixed typo in 'childern
user_input_scaled = scaler.transform(user_input)
prediction = model.predict(user_input_scaled)[0]
if prediction == 1:
    print("Prediction: More likely to claim Insurance")
else:
    print("Prediction: Less likely to claim Insurance")
```

```python
#Insurance riks assessment model with random forest classifier

from sklearn.model_selection import train_test_split#(for traing and testing)
from sklearn.ensemble import RandomForestClassifier#(for prediction)
from sklearn.preprocessing import StandardScaler#(for filter normalization)
from imblearn.over_sampling import SMOTE#(over evaluation for minority data, for handling class imbalance)
import numpy as np#(for numerical data handling)
X = data.drop(['charges','insuranceclaim'], axis =1)#(import data except these two rows)
y = data['insuranceclaim']#(output)
scaler = StandardScaler()#(initialization of scaler)
X_scaled = scaler.fit_transform(X)#(fit scaler into X data)
smote = SMOTE(random_state=42)#(to evluate Over spaling)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)#(minority data initilization)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size= 0.2, random_state=42)
model = RandomForestClassifier(random_state=42)#(using RFC)
model.fit(X_train, y_train)#(train the model using training data)
print("Insurance claim Prediction")#(input from user)
age = float(input("Enter age in years: "))
sex = int(input("Enter gender, 0 = female, 1 = male: "))
bmi = float(input("Enter bmi: "))
children = int(input("Enter number of children: "))
smoker = int(input("Enter smoking status, 0 = No, 1 = Yes: "))
region = int(input("Enter region: 0 = Northeast, 1 = northwest, 2 = southeast, 4 = southwest: "))
user_input = pd.DataFrame([[age, sex, bmi, children, smoker, region]], columns=['age', 'sex', 'bmi', 'children', 'smoker', 'region'])  # Fixed: column= -> columns= and fixed typo in 'childern
user_input_scaled = scaler.transform(user_input)#(using same scaler using during for training)
prediction = model.predict(user_input_scaled)[0]#(using prediction with randomforest class model)
if prediction == 1:
    print("Prediction: More likely to claim Insurance")
else:
    print("Prediction: Less likely to claim Insurance")
```
