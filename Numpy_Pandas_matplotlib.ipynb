```python
import numpy as np

arr1 = np.arange(1, 21)
print(np.sum(arr1), np.mean(arr1), np.median(arr1), np.std(arr1))
print(np.where(arr1 > 10))

arr2 = np.arange(1, 17).reshape(4, 4)
print(arr2)
print(arr2.T)
print(arr2.sum(axis=1))
print(arr2.sum(axis=0))

a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))
print(a + b)
print(a - b)
print(a * b)
print(np.dot(a, b))

arr3 = np.arange(12).reshape(3, 4)
print(arr3[:2, -2:])
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print(df.head())
print(df[['Age', 'Salary']].describe())
print(df[df['Department'] == 'HR']['Salary'].mean())

df['Bonus'] = df['Salary'] * 0.1

print(df[(df['Age'] >= 25) & (df['Age'] <= 30)])

print(df.groupby('Department')['Salary'].mean())

df.sort_values(by='Salary').to_csv('sorted_salary.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color=['red', 'blue', 'green', 'purple'])
plt.title("Marks")
plt.show()

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1, 0, 0, 0]
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.show()

data = np.random.randint(1, 101, 1000)
plt.hist(data)
plt.show()
```
