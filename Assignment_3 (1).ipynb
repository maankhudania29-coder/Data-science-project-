```python
import random
import numpy as np
import matplotlib.pyplot as plt

heads = 0
tails = 0
for _ in range(10000):
    if random.choice(['H', 'T']) == 'H':
        heads += 1
    else:
        tails += 1
print(heads/10000, tails/10000)

count = 0
trials = 10000
for _ in range(trials):
    if random.randint(1,6) + random.randint(1,6) == 7:
        count += 1
print(count/trials)

def prob_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        if 6 in [random.randint(1,6) for _ in range(10)]:
            success += 1
    return success / trials

print(prob_at_least_one_six())

colors = ['R']*5 + ['G']*7 + ['B']*8
prev = None
blue_count = 0
red_given_blue = 0

for _ in range(1000):
    curr = random.choice(colors)
    if prev == 'B':
        blue_count += 1
        if curr == 'R':
            red_given_blue += 1
    prev = curr

print(red_given_blue/blue_count if blue_count else 0)

sample = np.random.choice([1,2,3], size=1000, p=[0.25,0.35,0.4])
print(np.mean(sample), np.var(sample), np.std(sample))

exp_data = np.random.exponential(scale=5, size=2000)
plt.hist(exp_data, density=True)
x = np.linspace(0, max(exp_data), 1000)
y = (1/5)*np.exp(-x/5)
plt.plot(x, y)
plt.show()

uniform_data = np.random.uniform(0,1,10000)
sample_means = []

for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(uniform_data)
plt.show()

plt.hist(sample_means)
plt.show()
```
